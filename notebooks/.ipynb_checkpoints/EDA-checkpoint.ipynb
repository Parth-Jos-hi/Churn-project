```python
df.drop('customerID', axis = 1, inplace = True)
```

```python
for col in df.select_dtypes('object'):
    print(col, df[col].unique()[:10])
```

```python
df.dtypes
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].astype(int)
df['Churn'] = df['Churn'].astype('category')
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Apply string cleaning only to object columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].map(lambda x: x.strip().title() if isinstance(x, str) else x)

# Replace specific values
df.replace({'No Internet Service': 'No', 'No Phone Service': 'No'}, inplace=True)
```

```python
df['Churn'].value_counts(normalize=True)
```

```python
import math
# Select numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
num_cols = len(numeric_cols)

# Calculate number of rows needed (2 plots per row)
n_rows = math.ceil(num_cols / 2)

# Create subplots
fig, axes = plt.subplots(n_rows, 2, figsize=(7, 2.5 * n_rows))
axes = axes.flatten()  # Flatten to 1D for easy iteration

# Plot each numeric column
for i, col in enumerate(numeric_cols):
    sns.histplot(df[col], kde=True, bins=20, ax=axes[i], color='mediumseagreen')
    axes[i].set_title(col, fontsize=9)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')
    axes[i].tick_params(axis='both', labelsize=6)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

```python
import math
object_cols = df.select_dtypes(include='object').columns
num_objs = len(object_cols)
n_rows = math.ceil(num_objs / 2)
fig, axes = plt.subplots(n_rows, 2, figsize=(7, 2.5 * n_rows))
axes = axes.flatten()
for i, col in enumerate(object_cols):
    sns.countplot(data=df, x=col, ax=axes[i], palette='pastel', hue='Churn')  # Capital C
    axes[i].set_title(col, fontsize=9)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')
    axes[i].tick_params(axis='x', labelrotation=90, labelsize=6)
    axes[i].tick_params(axis='y', labelsize=6)
for j in range(i + 1, len(axes)):
    axes[j].axis('off')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Features to plot
features = ['MonthlyCharges', 'TotalCharges', 'tenure']

# Create subplots for boxplots
fig, axes = plt.subplots(1, 3, figsize=(12, 3))

# Plot each numeric feature against Churn
for i, col in enumerate(features):
    sns.boxplot(data=df, x='Churn', y=col, ax=axes[i], hue='Churn', palette='Set2', legend=False)
    axes[i].set_title(f'{col} vs Churn', fontsize=10)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')
    axes[i].tick_params(axis='x', labelsize=8)
    axes[i].tick_params(axis='y', labelsize=8)

plt.tight_layout()
plt.show()

plt.figure(figsize=(4, 3))
sns.countplot(data=df, x='Churn', hue='Churn', palette='pastel', legend=False)
plt.title('Churn Distribution', fontsize=10)
plt.xlabel('')
plt.ylabel('')
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.tight_layout()
plt.show()

```

### Categorical vs Target (Churn)

```python
# Standardize all column names to title case
df.columns = df.columns.str.strip().str.title()

# Now define your categorical columns
cat_cols = df.select_dtypes(include='object').columns

# Plot each categorical column vs Churn
for col in cat_cols:
    plt.figure(figsize=(5, 3))
    sns.countplot(x=col, hue='Churn', data=df)
    plt.title(f"{col} vs Churn")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
for col in cat_cols:
    churn_rate = df.groupby(col)['Churn'].value_counts(normalize=True).unstack().fillna(0)['Yes'] * 100
    churn_rate.sort_values(ascending=False).plot(kind='bar', figsize=(5,3))
    plt.title(f"Churn Rate (%) by {col}")
    plt.ylabel('Churn %')
    plt.show()
```
